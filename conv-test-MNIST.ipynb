```python
%load_ext autoreload
%autoreload 2

from FinalFlow import *
```

```python
train_path = 'data/mnist_train.csv'
```

```python
img_data = Image(path = train_path,test_frac = 0.05,val_frac = 0.94,image_size=(28,28), colour = 'gray')
```

```python
train_data = img_data.train_data
train_labels = img_data.train_labels

test_data = img_data.train_data
test_labels = img_data.train_labels
```

```python
%load_ext autoreload
%autoreload 2
%matplotlib notebook

Layer1 = Conv(X = np.array(train_data),in_channels = 1, out_channels = 5, kernel_size=3, stride=1, padding= 2  , Activation_fn = 'relu')
Layer2 = MaxPool(Layer1.output(),kernel_size=3, stride=2, padding=0)
Layer3 = Flatten(Layer2.output())
Layer4 = Linear(Input_Matrix = Layer3.output().T,Output_Dimension = 100,Activation_fn = 'relu')
Layer5 = Linear(Input_Matrix = Layer4.output(),Output_Dimension = 10,Activation_fn = 'relu')
Layer6 = Linear(Input_Matrix = Layer5.output(),Output_Dimension = 10,Activation_fn = 'softmax')
my_Model = Model([Layer1,Layer2,Layer3,Layer4,Layer5,Layer6])
my_Loss = my_Model.Loss(np.array(train_labels), Layer6.output()).CrossEntropy()
```

```python
# Parameters
learning_rate = 0.2
training_epochs = 200
epsilon = 0.01
opt = my_Model.Optimization(training_epochs, my_Loss, learning_rate, epsilon ).GradientDescent()
```

```python
yhat = my_Model.inference(Test_Inputs = train_data)
img_test = train2img(train_data[5,:])
print(img_test.shape)
plt.imshow(img_test)
```
